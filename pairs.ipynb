```python
# -*- coding=utf-8 -*-
import csv
res = []
#csv_file = ("perfume_top_score9.csv",encoding="utf-8" )#打开csv文件
with open('perfume_top_score9.csv', 'r',encoding="utf-8") as f:
    lines = csv.reader(f)
    for line in lines:
        # print(line)
        while '' in line:
            line.remove('')
        # print(line)

        for i in range(len(line)):
            for j in range(i+1, len(line)):
                tmp = []
                tmp.append(line[i])
                tmp.append(line[j])
                res.append(tmp)

    print(res)
with open('new_top_9.csv', "w", encoding="utf-8") as file:
    writer = csv.writer(file, lineterminator='\n')
    writer.writerows(res)
```
